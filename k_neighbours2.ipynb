```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

np.random.seed(42)
sns.set(font_scale=1.3)
```

```python
from sklearn.datasets import load_iris

raw_data = load_iris()
raw_data.keys()
```

```python
data = raw_data.copy()
target = data['target']
data_iris = data['data']
```

```python
indeksy = [2,3]
data_iris = data_iris[:,indeksy]
```

```python
print('data shape:', data_iris.shape)
print('target shape:', target.shape)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data_iris[:, 0], data_iris[:, 1], c=target, cmap='viridis')
plt.title('Wykres punktowy')
plt.xlabel('cecha_1: sepal_length')
plt.ylabel('cecha_2: sepal_width')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(data_iris, target)
```

```python
x_min, x_max = data_iris[:, 0].min() - 0.5, data_iris[:, 0].max() + 0.5
y_min, y_max = data_iris[:, 1].min() - 0.5, data_iris[:, 1].max() + 0.5

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
mesh = np.c_[xx.ravel(), yy.ravel()]
Z = classifier.predict(mesh)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 8))
plt.pcolormesh(xx, yy, Z, cmap='gnuplot', alpha=0.1)
plt.scatter(data_iris[:, 0], data_iris[:, 1], c=target, cmap='gnuplot', edgecolors='r')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title('3-class classification k=5')
plt.show()
```
